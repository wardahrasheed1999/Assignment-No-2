```python
import pandas as pd
```

```python
data = pd.read_csv("Pokemon.csv")
```

```python
data.isnull().sum()
```

## ADD COLUMS 
## Speed_High_Low

```python
 speed_mean = data["Speed"].mean()
def set_speed(val):
    if val < speed_mean:
        return "Speed Low"
    elif val == speed_mean:
        return "Speed neutral"
    else:
        return "Speed High"
```

```python
data["Speed_high_low"] = data["Speed"].apply(set_speed)
data
```

## ADD COLUMN
## HP_HIGH_LOW

```python
data["HP"].mean()
```

```python
hp_mean = data["HP"].mean()
def set_mean(val):
    if val < hp_mean:
        return "Hp Low"
    elif val == hp_mean:
        return "Hp Neutral"
    else:
        return "Hp High"
```

```python
data["Hp_High_Low"] = data["HP"].apply(set_mean)
data
```
